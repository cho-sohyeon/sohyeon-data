```python
driver = webdriver.Chrome()
driver.get('https://ohou.se/')
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from bs4 import BeautifulSoup as bs

# 검색창 위치 찾아서 검색하기
search = driver.find_element(By.XPATH, '//*[@id="global-search-combobox"]/div/input')
search.click()
search.send_keys('광파오븐' + "\n")
time.sleep(2)  # ← 없으면 페이지 로드 전에 다음 코드 실행됨

# 커뮤니티 탭 클릭하기
driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/header/div[2]/div/div/div[1]/nav/span[5]/div/a').click()
time.sleep(2)  # ← 없으면 페이지 로드 전에 다음 코드 실행됨

# 객체화
soup = bs(driver.page_source, 'html.parser')

# ❌ 문제 1: driver.find_elements는 Selenium 객체 반환 → soup이 아닌 driver로 찾아야 함
# ❌ 문제 2: url_list 변수가 없는데 url_list_result에서 참조함
links = driver.find_elements(By.CSS_SELECTOR, 'a[href*="/community/posts/"]')

# ✅ Selenium 객체에서 href 뽑기 (.get_attribute 써야 함)
url_list_result = list(set([
    a.get_attribute("href").split("?")[0]
    for a in links
    if a.get_attribute("href")
]))

print(f"수집된 URL: {len(url_list_result)}개")
url_list_result
```

```python
"""
오늘의집 커뮤니티 크롤러
- 검색어로 커뮤니티 검색 → 무한스크롤로 게시글 URL 전체 수집
- 각 게시글 본문 + 댓글(페이지네이션) 수집
- CSV 저장

실행 전 설치:
    pip install selenium beautifulsoup4 pandas tqdm
"""

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
import pandas as pd
import time
import json

# ── ⚙️ 설정 ──────────────────────────────────────────────────────────────────
KEYWORD   = "광파오븐"
SAVE_PATH = "ohou_광파오븐.csv"
HEADLESS  = False
# ─────────────────────────────────────────────────────────────────────────────


# ── STEP 1: 드라이버 실행 ─────────────────────────────────────────────────────
def init_driver():
    options = Options()
    if HEADLESS:
        options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("lang=ko_KR")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    )
    driver = webdriver.Chrome(options=options)
    driver.implicitly_wait(3)
    driver.set_page_load_timeout(20)
    return driver


# ── STEP 2: 무한스크롤로 게시글 URL 전체 수집 ────────────────────────────────
def get_url_list(driver, keyword):
    print(f"\n[1단계] '{keyword}' 검색 결과 URL 수집 중...")

    # 검색 결과 페이지 직접 접근
    search_url = f"https://ohou.se/search/community?query={keyword}&search_affect_type=History"
    try:
        driver.get(search_url)
    except Exception:
        pass
    time.sleep(2)

    # 무한스크롤 - 새 게시글이 안 나올 때까지 반복
    prev_count = 0
    stall      = 0
    pbar = tqdm(desc="  스크롤", unit="개", dynamic_ncols=True)

    while stall < 4:
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
        time.sleep(1.5)
        cur_count = len(driver.find_elements(By.CSS_SELECTOR, 'a[href*="/community/posts/"]'))
        if cur_count > prev_count:
            pbar.update(cur_count - prev_count)
            stall = 0
        else:
            stall += 1
        prev_count = cur_count
    pbar.close()

    # URL + 제목 수집 (중복 제거)
    links = driver.find_elements(By.CSS_SELECTOR, 'a[href*="/community/posts/"]')
    seen, url_list = set(), []

    for a in links:
        href  = (a.get_attribute("href") or "").split("?")[0]
        title_els = a.find_elements(By.TAG_NAME, "h3")
        title = title_els[0].text.strip() if title_els else ""

        if href and href not in seen:
            seen.add(href)
            url_list.append({"url": href, "title": title})

    print(f"  ✅ 총 {len(url_list)}개 URL 수집 완료\n")
    return url_list


# ── STEP 3: 게시글 본문 파싱 (__NEXT_DATA__ 기반) ────────────────────────────
def parse_body(soup):
    nd_tag = soup.find("script", id="__NEXT_DATA__")
    if not nd_tag:
        return "", "", "", "", 0, 0, 0

    try:
        nd      = json.loads(nd_tag.string)
        queries = (nd.get("props", {})
                     .get("pageProps", {})
                     .get("dehydratedState", {})
                     .get("queries", []))
    except Exception:
        return "", "", "", "", 0, 0, 0

    post_data, cmt_data = {}, {}
    for q in queries:
        key = q.get("queryKey", [])
        sd  = q.get("state", {}).get("data", {})
        if len(key) == 3 and key[:2] == ["community", "posts"]:
            post_data = sd.get("post", {})
        if len(key) >= 4 and "comments" in key:
            cmt_data = sd

    # 본문 텍스트 조립
    body_parts = []
    for comp in (post_data.get("components") or []):
        if comp.get("componentType") == "TEXT":
            t = comp.get("content") or comp.get("description") or ""
            if t.strip():
                body_parts.append(t.strip())
    body = "\n".join(body_parts)

    # 작성자
    writer_obj = post_data.get("writer") or {}
    profile    = writer_obj.get("profile") or writer_obj
    writer     = profile.get("nickname", "") if isinstance(profile, dict) else ""

    title      = post_data.get("title", "")
    created_at = (post_data.get("createdAt") or "")[:10]
    like_count = post_data.get("likeCount", 0)
    view_count = post_data.get("viewCount", 0)
    cmt_total  = cmt_data.get("totalWithChildCount", 0)

    return title, body, writer, created_at, like_count, view_count, cmt_total


# ── STEP 4: 댓글 파싱 (__NEXT_DATA__ 기반) ───────────────────────────────────
def parse_comments(soup):
    nd_tag = soup.find("script", id="__NEXT_DATA__")
    if not nd_tag:
        return []
    try:
        nd      = json.loads(nd_tag.string)
        queries = (nd.get("props", {})
                     .get("pageProps", {})
                     .get("dehydratedState", {})
                     .get("queries", []))
    except Exception:
        return []

    comments = []
    for q in queries:
        if "comments" not in q.get("queryKey", []):
            continue
        for c in q.get("state", {}).get("data", {}).get("comments", []):
            writer = (c.get("writer") or {}).get("nickname", "")
            text   = (c.get("text")   or "").strip()
            date   = (c.get("createdAt") or "")[:10]
            if text:
                comments.append(f"[{writer} / {date}] {text}")
            # 대댓글
            for r in (c.get("replies") or []):
                rw   = (r.get("writer") or {}).get("nickname", "")
                rtxt = (r.get("text")   or "").strip()
                rdate = (r.get("createdAt") or "")[:10]
                if rtxt:
                    comments.append(f"  └[{rw} / {rdate}] {rtxt}")
    return comments


# ── STEP 5: 댓글 페이지네이션 버튼 클릭 → 전체 댓글 수집 ─────────────────────
def get_all_comments(driver):
    """
    지식인 nextPageButton 방식과 동일:
    댓글 > 버튼 반복 클릭 → 모든 페이지 댓글 수집
    """
    all_comments = []

    while True:
        # 페이지 끝까지 스크롤
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
        time.sleep(0.8)

        # 로그인 페이지 감지 → 즉시 중단
        if "sign_in" in driver.current_url or "login" in driver.current_url:
            driver.back()
            time.sleep(1)
            break

        # 현재 페이지 댓글 수집
        soup = bs(driver.page_source, "html.parser")
        all_comments.extend(parse_comments(soup))

        # 다음 페이지(>) 버튼 찾기
        next_btn = None
        try:
            btns = driver.find_elements(By.CSS_SELECTOR,
                '[class*="pagination"] button, [class*="Pagination"] button, '
                '[class*="paging"] button, nav button'
            )
            if btns:
                next_btn = btns[-1]   # 마지막 버튼 = > 버튼
        except Exception:
            pass

        if not next_btn:
            break

        # disabled 확인 (마지막 페이지면 비활성화)
        try:
            is_disabled = (
                not next_btn.is_enabled() or
                next_btn.get_attribute("disabled") is not None or
                "disabled" in (next_btn.get_attribute("class") or "")
            )
            if is_disabled:
                break
        except Exception:
            break

        # 버튼 클릭
        try:
            driver.execute_script("arguments[0].scrollIntoView({block:'center'});", next_btn)
            time.sleep(0.3)
            driver.execute_script("arguments[0].click();", next_btn)
            time.sleep(1)

            if "sign_in" in driver.current_url or "login" in driver.current_url:
                driver.back()
                time.sleep(1)
                break
        except Exception:
            break

    return all_comments


# ── STEP 6: 게시글 1개 크롤링 ────────────────────────────────────────────────
def crawl_post(driver, url, keyword=""):
    try:
        driver.get(url)
    except Exception:
        pass
    time.sleep(1)

    # 댓글 전체 수집 (페이지네이션 버튼 클릭)
    all_comments = get_all_comments(driver)

    # 최종 페이지 소스 파싱
    soup  = bs(driver.page_source, "html.parser")
    title, body, writer, created_at, like, view, cmt_total = parse_body(soup)

    # 키워드 필터 - 제목·본문에 키워드 없으면 스킵
    if keyword and keyword not in title and keyword not in body:
        tqdm.write(f"  ⏭ 스킵 (키워드 없음): {title[:40]}")
        return None

    return {
        "제목"   : title,
        "본문"   : body,
        "작성자" : writer,
        "날짜"   : created_at,
        "좋아요" : like,
        "조회수" : view,
        "댓글수" : cmt_total,
        "댓글"   : "\n".join(all_comments),
        "URL"    : url,
    }


# ── 메인 ─────────────────────────────────────────────────────────────────────
if __name__ == "__main__":

    driver = init_driver()

    # 1단계: URL 수집
    url_info_list = get_url_list(driver, KEYWORD)

    # 2단계: 게시글 본문 + 댓글 수집
    print(f"[2단계] 게시글 크롤링 중... (총 {len(url_info_list)}개)")
    df = pd.DataFrame(columns=["제목", "본문", "작성자", "날짜", "좋아요", "조회수", "댓글수", "댓글", "URL"])

    for idx, info in enumerate(tqdm(url_info_list, desc="게시글 크롤링", unit="개")):
        try:
            data = crawl_post(driver, info["url"], keyword=KEYWORD)
            if data is None:
                continue
            df.loc[len(df)] = data
            tqdm.write(f"  ✔ [{idx+1}/{len(url_info_list)}] {data['제목'][:40]}  댓글 {len(data['댓글'].splitlines())}개")
        except Exception as e:
            tqdm.write(f"  ⚠ 오류 [{idx+1}] {info['url'][:50]}: {e}")
        time.sleep(0.5)

    driver.quit()

    # 3단계: 저장
    df.to_csv(SAVE_PATH, index=False, encoding="utf-8-sig")
    print(f"\n✅ 완료! '{SAVE_PATH}' 저장 ({len(df)}개 게시글)")
    
```

```python
"""
오늘의집 커뮤니티 크롤러 최종버전
- 드라이버 창에서 직접 네이버 로그인 후 크롤링
- 검색어로 커뮤니티 게시글 URL 전체 수집 (무한스크롤)
- 각 게시글 본문 + 댓글 수집
- CSV 저장

실행 전 설치:
    pip install selenium beautifulsoup4 pandas tqdm
"""

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
import pandas as pd
import time
import json

# ── ⚙️ 설정 ──────────────────────────────────────────────────────────────────
KEYWORD   = "광파오븐"
SAVE_PATH = "ohou_광파오븐.csv"
SCROLL_PAUSE = 3      # 스크롤 대기 (높을수록 차단 위험 낮음)
PAGE_PAUSE   = 2      # 게시글 이동 대기
# ─────────────────────────────────────────────────────────────────────────────


# ── STEP 1: 드라이버 실행 + 로그인 대기 ──────────────────────────────────────
def init_driver():
    options = Options()
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("lang=ko_KR")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    )
    driver = webdriver.Chrome(options=options)
    driver.implicitly_wait(3)
    driver.set_page_load_timeout(20)

    # 오늘의집 접속 후 직접 로그인 대기
    driver.get('https://ohou.se/')
    print("=" * 50)
    print("👉 지금 뜬 크롬 창에서 네이버 로그인 해주세요!")
    print("   로그인 완료되면 여기서 엔터를 누르세요.")
    print("=" * 50)
    input()
    print("✅ 로그인 확인! 크롤링 시작합니다.\n")

    return driver


# ── STEP 2: 무한스크롤로 게시글 URL 전체 수집 ────────────────────────────────
def get_url_list(driver, keyword):
    print(f"[1단계] '{keyword}' 검색 결과 URL 수집 중...")

    try:
        driver.get(f"https://ohou.se/search/community?query={keyword}&search_affect_type=History")
    except Exception:
        pass
    time.sleep(3)

    seen = set()
    url_list = []
    stall = 0

    pbar = tqdm(desc="  스크롤 수집", unit="개", dynamic_ncols=True)

    while stall < 6:
        # 현재 보이는 링크 즉시 저장 (가상DOM 대응)
        links = driver.find_elements(By.TAG_NAME, 'a')
        added = 0
        for a in links:
            href = a.get_attribute('href') or ''
            if 'contents.ohou.se/community/posts/' in href:
                clean = href.split('?')[0]
                if clean not in seen:
                    seen.add(clean)
                    url_list.append(clean)
                    added += 1

        if added > 0:
            pbar.update(added)
            stall = 0
        else:
            stall += 1

        # 스크롤 내리기
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(SCROLL_PAUSE)

    pbar.close()
    print(f"  ✅ 총 {len(url_list)}개 URL 수집 완료\n")
    return url_list


# ── STEP 3: 게시글 본문 파싱 ─────────────────────────────────────────────────
def parse_body(soup):
    nd_tag = soup.find("script", id="__NEXT_DATA__")
    if not nd_tag:
        return "", "", "", "", 0, 0, 0
    try:
        nd      = json.loads(nd_tag.string)
        queries = (nd.get("props", {})
                     .get("pageProps", {})
                     .get("dehydratedState", {})
                     .get("queries", []))
    except Exception:
        return "", "", "", "", 0, 0, 0

    post_data, cmt_data = {}, {}
    for q in queries:
        key = q.get("queryKey", [])
        sd  = q.get("state", {}).get("data", {})
        if len(key) == 3 and key[:2] == ["community", "posts"]:
            post_data = sd.get("post", {})
        if len(key) >= 4 and "comments" in key:
            cmt_data = sd

    body_parts = []
    for comp in (post_data.get("components") or []):
        if comp.get("componentType") == "TEXT":
            t = comp.get("content") or comp.get("description") or ""
            if t.strip():
                body_parts.append(t.strip())
    body = "\n".join(body_parts)

    writer_obj = post_data.get("writer") or {}
    profile    = writer_obj.get("profile") or writer_obj
    writer     = profile.get("nickname", "") if isinstance(profile, dict) else ""

    title      = post_data.get("title", "")
    created_at = (post_data.get("createdAt") or "")[:10]
    like_count = post_data.get("likeCount", 0)
    view_count = post_data.get("viewCount", 0)
    cmt_total  = cmt_data.get("totalWithChildCount", 0)

    return title, body, writer, created_at, like_count, view_count, cmt_total


# ── STEP 4: 댓글 파싱 ────────────────────────────────────────────────────────
def parse_comments(soup):
    nd_tag = soup.find("script", id="__NEXT_DATA__")
    if not nd_tag:
        return []
    try:
        nd      = json.loads(nd_tag.string)
        queries = (nd.get("props", {})
                     .get("pageProps", {})
                     .get("dehydratedState", {})
                     .get("queries", []))
    except Exception:
        return []

    comments = []
    for q in queries:
        if "comments" not in q.get("queryKey", []):
            continue
        for c in q.get("state", {}).get("data", {}).get("comments", []):
            writer = (c.get("writer") or {}).get("nickname", "")
            text   = (c.get("text")   or "").strip()
            date   = (c.get("createdAt") or "")[:10]
            if text:
                comments.append(f"[{writer} / {date}] {text}")
            for r in (c.get("replies") or []):
                rw    = (r.get("writer") or {}).get("nickname", "")
                rtxt  = (r.get("text")   or "").strip()
                rdate = (r.get("createdAt") or "")[:10]
                if rtxt:
                    comments.append(f"  └[{rw} / {rdate}] {rtxt}")
    return comments


# ── STEP 5: 댓글 페이지네이션 버튼으로 전체 댓글 수집 ────────────────────────
def get_all_comments(driver):
    all_comments = []

    while True:
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
        time.sleep(1)

        # 로그인 페이지 감지
        if "sign_in" in driver.current_url or "login" in driver.current_url:
            driver.back()
            time.sleep(2)
            break

        # 현재 페이지 댓글 수집
        soup = bs(driver.page_source, "html.parser")
        all_comments.extend(parse_comments(soup))

        # 다음 페이지(>) 버튼 찾기
        next_btn = None
        try:
            btns = driver.find_elements(By.CSS_SELECTOR,
                '[class*="pagination"] button, [class*="Pagination"] button, '
                '[class*="paging"] button, nav button'
            )
            if btns:
                next_btn = btns[-1]
        except Exception:
            pass

        if not next_btn:
            break

        try:
            is_disabled = (
                not next_btn.is_enabled() or
                next_btn.get_attribute("disabled") is not None or
                "disabled" in (next_btn.get_attribute("class") or "")
            )
            if is_disabled:
                break
        except Exception:
            break

        try:
            driver.execute_script("arguments[0].scrollIntoView({block:'center'});", next_btn)
            time.sleep(0.5)
            driver.execute_script("arguments[0].click();", next_btn)
            time.sleep(1.5)
            if "sign_in" in driver.current_url or "login" in driver.current_url:
                driver.back()
                time.sleep(2)
                break
        except Exception:
            break

    return all_comments


# ── STEP 6: 게시글 1개 크롤링 ────────────────────────────────────────────────
def crawl_post(driver, url, keyword=""):
    try:
        driver.get(url)
    except Exception:
        pass
    time.sleep(PAGE_PAUSE)

    all_comments = get_all_comments(driver)

    soup  = bs(driver.page_source, "html.parser")
    title, body, writer, created_at, like, view, cmt_total = parse_body(soup)

    # 키워드 필터
    if keyword and keyword not in title and keyword not in body:
        tqdm.write(f"  ⏭ 스킵: {title[:40]}")
        return None

    return {
        "제목"   : title,
        "본문"   : body,
        "작성자" : writer,
        "날짜"   : created_at,
        "좋아요" : like,
        "조회수" : view,
        "댓글수" : cmt_total,
        "댓글"   : "\n".join(all_comments),
        "URL"    : url,
    }


# ── 메인 ─────────────────────────────────────────────────────────────────────
if __name__ == "__main__":

    # 1단계: 드라이버 실행 + 로그인
    driver = init_driver()

    # 2단계: URL 수집
    url_list = get_url_list(driver, KEYWORD)

    # 3단계: 게시글 본문 + 댓글 수집
    print(f"[2단계] 게시글 크롤링 중... (총 {len(url_list)}개)")
    df = pd.DataFrame(columns=["제목", "본문", "작성자", "날짜", "좋아요", "조회수", "댓글수", "댓글", "URL"])

    for idx, url in enumerate(tqdm(url_list, desc="게시글 크롤링", unit="개")):
        try:
            data = crawl_post(driver, url, keyword=KEYWORD)
            if data is None:
                continue
            df.loc[len(df)] = data
            tqdm.write(f"  ✔ [{idx+1}/{len(url_list)}] {data['제목'][:40]}  댓글 {len(data['댓글'].splitlines())}개")
        except Exception as e:
            tqdm.write(f"  ⚠ 오류 [{idx+1}] {url[:50]}: {e}")
        time.sleep(0.5)

    driver.quit()


    # 4단계: 저장
    df.to_csv(SAVE_PATH, index=False, encoding="utf-8-sig")
    print(f"\n✅ 완료! '{SAVE_PATH}' 저장 ({len(df)}개 게시글)")
```
