## 5. 통합 코드 작성하기

```python
from tqdm import tqdm
```

```python
driver = webdriver.Chrome()
```

```python
df = pd.DataFrame(columns= ['제목', '본문' , '답변']) #최종 데이터를 담을 빈 데이터프레임 생성하기

for index, url  in enumerate(tqdm(url_list_result)): 
    driver.get(url)
    time.sleep(1) #화면로딩시간 

    
    #답변더보기 클릭
    try:
        while True:
            time.sleep(1)
            body = driver.find_element(By.CSS_SELECTOR, 'body')
            body.send_keys(Keys.END)
            driver.find_element(By.ID, 'nextPageButton').click()
            time.sleep(1)
    except:
        pass

    
    #객체화
    soup = bs(driver.page_source, 'html.parser')
    
    #제목
    title = soup.select_one('div.endTitleSection').text
    title = title.replace('\n', '') #전처리과정
    title = title.replace('\t', '') #전처리과정
    title = title[2:] #'질문' 문자열 삭제하기

    #본문
    contents = soup.select_one('div.questionDetail').text

    #답변
    return_list = soup.select('div.se-main-container')

    final_return = []
    for i in return_list:
        final_return.append(i.text)


    #최종 데이터 정리
    data = [title, contents, final_return]
    df.loc[index] = data
```

```python
print(df.shape)   # (2800, 3) 확인
df.head()         # 앞 5행 미리보기
```

```python
df.to_csv('광파오븐_지식인.csv', index=False, encoding='utf-8-sig')
print(f"저장 완료! 총 {len(df)}행")
```
